```python
import os

current_path=os.getcwd()
final_data_path=os.path.join(current_path,"final_data")

tif_file_path=[]
shp_file_path=[]
def list_files_recursively(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
          
            if "evi" in file.lower():
                continue
            if "ndmi" in file.lower():
                continue
            if "ndvi" in file.lower():
                continue

            if (".tif" in file.lower()) and (".aux.xml" not in file.lower()):
                tif_file_path.append(file)
                continue
            if  (".aux.xml" not in file.lower()):
                shp_file_path.append(file)

# Call the function with the current directory
list_files_recursively('./final_data')
```

```python
temp_tif=[]
temp_shp=[]
for path in tif_file_path:
    file_name=path.split("\\")[-1]
    if file_name in temp_tif:
        continue
    else:
        temp_tif.append(path)

for path in shp_file_path:
    file_name=path.split("\\")[-1]
    if file_name in temp_shp:
        continue
    else:
        temp_shp.append(path)
```

```python
temp_shp = [x for x in temp_shp if ".json" not in x]
```

```python
categories = ['unidentifiable','bamboo','forest','rice_field']
```

```python
def check_enough_categories(tif_file_path, shp_file_path, categories):
    missing_files = []
    for file in tif_file_path:
        for category in categories:
            concat_str = file.split(".")[0] + "_" + category + ".shp"
            if concat_str not in shp_file_path:
                print(file)
                missing_files.append(file)

    return missing_files
```
